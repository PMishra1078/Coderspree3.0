```python
# Extract features using refined regular expressions
# Extract entire processor names
# Extract entire processor names including values in and before the bracket
df['Processor'] = df['Feature'].str.extract(r'((Intel|AMD)[^\d]+ Core [^\d]+ Processor [^\d]+\([^)]+\))')
df['Processor'] = df['Processor'].str.strip()  # Remove leading/trailing spaces
df['RAM'] = df['Feature'].str.extract(r'(\d+ GB DDR4 RAM)')
df['Storage'] = df['Feature'].str.extract(r'(\d+ [^\d]+ SSD|\d+ [^\d]+ HDD)')
df['Operating_System'] = df['Feature'].str.extract(r'([^\d]+ Windows \d+ [^\d]+ Operating System)')
df['Display_Size'] = df['Feature'].str.extract(r'([\d.]+ [^\d]+ Display)')
df['Warranty'] = df['Feature'].str.extract(r'(\d+ Year [^\d]+Warranty)')
```

```python
# Display the first 5 rows of the resulting DataFrame
print(df[['Product', 'Processor', 'RAM', 'Storage', 'Operating_System', 'Display_Size', 'Warranty']].head())
```
